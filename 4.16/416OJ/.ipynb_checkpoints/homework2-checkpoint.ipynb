```python
import os, sys
import pandas as pd

path = "/Users/apple/Documents/Jupytor_Notebook/ai_programming/4.16/416OJ/ndb_total"
dirs = os.listdir( path )
al = []
```

```python
for file in dirs:
    try :
        s = file
        file = path + '/' + file
        Fi = pd.DataFrame(pd.read_json(file))
        st = Fi['report']['food']['name']
        df = pd.DataFrame(Fi['report']['food']['nutrients'])
        del df['measures']
        df['identify'] = st
        df['coder'] = s[:4]
        al += [df]
    except :
        pass
```

```python
new = pd.DataFrame(columns = ['derivation','group','name','nutrient_id','unit','value','identify','coder'])
search = input()
counter = 0
for i in range(len(al)) :
    m = al[i]
    if search in list(m.nutrient_id) :
        counter += 1
        new = new.append(m[m['nutrient_id'] == search])
new = new.sort_values(by = 'value')
new = new[::-1]
new[:20]
```
