## kNN

```python
from sklearn import datasets 
raw = datasets.load_iris()
raw.keys()
```

```python
from sklearn.cluster import KMeans
k_means = KMeans(n_clusters=3, random_state=0)
```

```python
import pandas as pd
from sklearn import manifold
tsne = manifold.TSNE(n_components=2,init='pca')
dt1 = pd.DataFrame(tsne.fit_transform(pd.DataFrame(raw.data)))
dt2 = pd.DataFrame(raw.target)
dt = pd.concat([dt2 , dt1],axis=1)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split 
x = np.array(dt.iloc[:,1:])
y = np.array(dt.iloc[:,0])
x_train, x_test, y_train, y_test = train_test_split(x , y, test_size = 0.1, random_state = 1)
```

```python
k_f = k_means.fit(x_train)
pre = k_f.predict(x_train)
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
s = pd.DataFrame(x_train).iloc[:,0]
t = pd.DataFrame(x_train).iloc[:,1]
sns.scatterplot(x=s , y=t , hue=pre)
plt.show()
```

```python
sns.scatterplot(x=pd.DataFrame(x_train).iloc[:,0] , y=pd.DataFrame(x_train).iloc[:,1] , hue=y_train)
plt.show()
```
