```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
import jieba
```

由于有换行符，所以需要strip来去除结尾的无关符号

```python
with open("train_texts.txt") as tr:
    train_lines=tr.readlines()
    for i in range(len(train_lines)):
        train_lines[i]=train_lines[i].strip()
```

```python
train_lines[:8]
```

每条新闻长度有限，所以不需要限制100词长度

同样需要对含换行符的标签处理

```python
with open("train_labels.txt") as train_label:
    tr_label_lines=train_label.readlines()
    train_labels=[]
    for i in tr_label_lines:
        train_labels.append(i.strip())
```

```python
train_labels
```

**分词**

```python
tfidf1=TfidfVectorizer(stop_words='english')
t_train=tfidf1.fit_transform(train_lines)
```

**朴素贝叶斯**

```python
from sklearn.naive_bayes import MultinomialNB
mod = MultinomialNB().fit(t_train,train_labels)
```

```python
with open("test_texts.txt") as test:
    test_lines=test.readlines()
    for i in range(len(test_lines)):
        test_lines[i]=test_lines[i].strip()
```

```python
test_lines[:8]
```

```python
with open("test_labels.txt") as test_label:
    test_label_lines=test_label.readlines()
    test_labels=[]
    for i in test_label_lines:
        test_labels.append(i.strip())
```

```python
test_labels
```

```python
t_test=tfidf1.transform(test_lines)
```

```python
pred=mod.predict(t_test)
```

```python
pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(test_labels , pred)
```

```python
err = 1-accuracy_score(test_labels , pred)
print(err)
```

主要任务在于调包。。。

**回归**

```python
from sklearn.linear_model import LogisticRegressionCV
lr_model = LogisticRegressionCV(solver='newton-cg', multi_class='multinomial', cv=10, n_jobs=-1)
lr_model.fit(t_train, train_labels)
acc = accuracy_score(test_labels , lr_model.predict(t_test))
```
