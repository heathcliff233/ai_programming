```python
import numpy as np
import pandas as pd
cups = [1,3,4,7]
weights = [50,65,60,63]
cup = np.array(cups)
weight = np.array(weights)
```

### method 1

```python
mean_cups = sum(cups)/len(cups)
mean_weights = sum(weights)/len(weights)
ups = sum([(cups[i] - mean_cups)*(weights[i] - mean_weights) for i in range(4)])
downs1 = sum([(i - mean_cups)**2 for i in cups])
downs2 = sum([(i - mean_weights)**2 for i in weights])
relate = ups/(downs1**0.5 * downs2**0.5)
relate
```

### method 2

```python
upper = sum((cup-cup.mean())*(weight-weight.mean()))
lower1 = sum((cup-cup.mean())*(cup-cup.mean()))
lower2 = sum((weight-weight.mean())*(weight-weight.mean()))
relativity = upper/(lower1**0.5 * lower2**0.5)
relativity
```

### method 3

```python
df = pd.DataFrame({'a':cup , 'b':weight})
df.corr()
```
